```python
cat_cols = {'CarName', 'fueltype', 'aspiration', 'carbody', 'enginetype', 'fuelsystem', 'drivewheel', 'enginelocation'}
rang_cols = {'symboling', 'doornumber', 'cylindernumber'}
num_cols = set(data.columns) - cat_cols - rang_cols
num_cols.remove('price')
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import Normalizer
```

## Baseline

Только числовые колонки

```python
X = data[num_cols]
y = data['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, shuffle=True, random_state=42)

transformer = Normalizer().fit(X_train)

reg = LinearRegression()

scores = cross_val_score(reg, transformer.transform(X_train), y_train, cv=5, scoring = 'neg_mean_absolute_error')
print((-1) * scores.mean())
preds = reg.fit(transformer.transform(X_train), y_train).predict(transformer.transform(X_test))
print(mean_absolute_error(preds, y_test))
```

Зафиксировали значения метрик.

    Для обучающей = 2717

## Обработка числовых колонок

Воспользуемся методом наименьших квадратов и оценим значимость числовых признаков

```python
resultsm = smf.ols(f'price ~ {"+".join(num_cols)}', data=data).fit()
resultsm.summary()
```

Будем убирать незначимые параметры по одному

```python
useful_cols = num_cols - set(['boreratio', 'carheight', 'carheight', 'carlength', 'citympg', 'wheelbase', 'highwaympg'])
```

```python
resultsm = smf.ols(f'price ~ {"+".join(useful_cols)}', data=data).fit()
resultsm.summary()
```

```python
X = data[useful_cols]
y = data['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, shuffle=True, random_state=42)

transformer = Normalizer().fit(X_train)

reg = LinearRegression()

scores = cross_val_score(reg, transformer.transform(X_train), y_train, cv=5, scoring = 'neg_mean_absolute_error')
print((-1) * scores.mean())
preds = reg.fit(transformer.transform(X_train), y_train).predict(transformer.transform(X_test))
print(mean_absolute_error(preds, y_test))
```

Значения ошибки и на тестовой, и на обучающей снизились, т.к. мы избавились от незначимых признаков
